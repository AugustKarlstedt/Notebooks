```julia
A = [4 0; 3 5]
```

```julia
transpose(A)
```

```julia
adjoint(A)
```

```julia
A'
```

Not sure what the difference is between transpose and adjoint... also julia calls adjoint when using the `'` postfix operator e.g. `A'`

```julia
A'*A
```

```julia
E = LinearAlgebra.eigen(A'*A)
```

```julia
E.values, E.vectors
```

```julia
S = E.values
```

```julia
S = [sqrt(s) for s in S]
```

```julia
sort!(S, rev=true)
```

```julia
S = diagm(0 => S)
```

```julia
S_inv = inv(S)
```

```julia
V = E.vectors
```

```julia
V
```

```julia
V'
```

```julia
U = A*V*S_inv
```

```julia
U*S*V'
```

## WE DID IT! We've computed the _singular values_ and then reconsructed our original matrix :)

Now let's try it with an image.

```julia
? open
```

```julia
open('plants.jpg')
```

```julia
using Images
```

```julia
image = Images.load("plants.jpg")
```

```julia
image[1, 1], image[end, end]
```

```julia
image = Gray.(image)
```

```julia
image[1, 1], image[end, end]
```

```julia
image_T = image'
```

```julia
image_T_image = image_T * image
```

```julia
m = min(size(image)...)
```

```julia
eigvecs(image)
```

```julia
eigens = eigen(image)
```

```julia
? eigen
```

```julia
S = eigens.values
```

```julia
sort!(S, rev=true)
```

```julia
S = S[1:m]
```

```julia
S = [sqrt(s) for s in S]
```

```julia
S = diagm(0 => S)
```

```julia
Gray.(S)
```

```julia
S_inv = inv(S)
```

```julia
Gray.(S_inv)
```

```julia
V = eigens.vectors[1:m, 1:m]
```

```julia
V'
```

```julia
U = image_T * V * S_inv
```

```julia
U*S*V'
```

WE GOT OUR ORIGINAL IMAGE BACK. THAT'S AMAZING!

What if reduce our number of singular values?

```julia
k = 960
```

```julia
U_new = U[:, 1:k]
```

```julia
size(U_new)
```

```julia
V_new = V[:, 1:k]
```

```julia
S_new = S[1:k, 1:k]
```

```julia
U_new*S_new*V_new'
```

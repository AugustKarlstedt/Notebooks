```python
def get_pair():
  u1 = torch.rand(1)
  u2 = torch.rand(1)

  R = torch.sqrt(-2*torch.torch.log(u1))
  theta = 2*math.pi*u2

  z0 = R * torch.cos(theta)
  z1 = R * torch.sin(theta)
  
  return z0, z1
```

```python
n_samples = 4096

points_x = []
points_y = []

for i in range(n_samples):
  x,y = get_pair()
  points_x.append(x.item())
  points_y.append(y.item())
```

```python
fig, ax = plt.subplots()
_ = ax.scatter(points_x, points_y, alpha=0.125)
```
